```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
training_data = np.array(
    [[0, 0], [0, 1], [1, 0], [1,1]], 'float32')
target_data = np.array([[0], [1], [1], [0]], 'float32')
```

###경사하강 알고리즘
https://shuuki4.github.io/deep%20learning/2016/05/20/Gradient-Descent-Algorithm-Overview.html

```python
model = Sequential()
model.add(Dense(32, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='mse', optimizer='adam', metrics=['binary_accuracy'])
model.summary()
```

```python
fit_hist = model.fit(training_data, target_data, epochs=500, verbose=1)
```

```python
plt.plot(fit_hist.history['loss'])
plt.show()
```

```python
inp = list(map(int, input().split()))
qwe = np.array(inp)
print('입력 값')
qwe = qwe.reshape(1, 2)
print('결과 값', model.predict(qwe)[0][0].round())
```
