```python
exam_data = {'이름':['경석','연주','선희'], #딕셔너리로 만들기
             '수학':[90,85,80],
             '영어':[99,98,97],
             '음악':[88,87,86],
             '체육':[79,78,77]}
df = pd.DataFrame(exam_data)
df.set_index('이름',inplace=True) #이름을 인덱스로 세팅
df
```
